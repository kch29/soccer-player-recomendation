```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_original = pd.read_csv("Diabetes.csv")
df_original.head()
```

```python
cols = [c for c in df_original.columns if c not in ["Pregancies", "Outcome"]]
df = df_original.copy()
df[cols] = df[cols].replace({0: np.NaN})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
pd.plotting.scatter_matrix(df, figsize=(7,7));
```

```python
df2 = pd.read_csv("height_weight.csv")
df2.info()
df2.describe()
```

```python
plt.hist2d(df2["height"], df2["weight"], bins=20)
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
hist, x_edge, y_edge = np.histogram2d(df2["height"], df2["weight"], bins=20)
x_center = 0.5 * (x_edge[1:] + x_edge[:-1])
y_center = 0.5 * (y_edge[1:] + y_edge[:-1])

plt.contour(x_center, y_center, hist, levels = 4)
plt.xlabel("Height")
plt.ylabel("Weight");
```

```python
sb.kdeplot(df2["height"], df2["weight"], cmap="viridis");
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap="magma", alpha=0.3);
```
